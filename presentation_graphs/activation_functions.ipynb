```python
import torch
```

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
activation_functions = {
    "ReLU": torch.nn.ReLU(),
    "Sigmoid": torch.nn.Sigmoid(),
    "Tanh": torch.nn.Tanh(),
    "SELU": torch.nn.SELU()
}

section = torch.Tensor(np.linspace(-10, 10, 200))
```

```python
for func_name, func in activation_functions.items():
    plt.plot(section.numpy(), func(section).numpy())
    plt.xlabel('X, input')
    plt.ylabel('Y, output')
    plt.title(func_name)
    plt.show()
```

```python
relu1 = lambda x: - (torch.nn.ReLU()(x)) + 4
relu2 = lambda x: (torch.nn.ReLU()(x - 2) * 2) - 2
relu3 = lambda x: (torch.nn.ReLU()(x - 4)) + 2
```

```python
section = torch.Tensor(np.linspace(-10, 10, 200))

plt.plot(section.numpy(), relu1(section).numpy(), label="- ReLU(X) + 4")
plt.plot(section.numpy(), relu2(section).numpy(), label="2 * ReLU(X - 2) - 2")
plt.plot(section.numpy(), relu3(section).numpy(), label="ReLU(X - 4) + 2")
plt.xlabel('X, input')
plt.ylabel('Y, output')
plt.legend()
plt.show()
```

```python
plt.plot(section.numpy(), (relu1(section) + relu2(section) + relu3(section)).numpy(), label="ReLU sum")
plt.xlabel('X, input')
plt.ylabel('Y, output')
plt.legend()
plt.show()
```
