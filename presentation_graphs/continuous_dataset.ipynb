```python
import numpy as np
import pandas as pd
```

```python
btc_to_usd = pd.read_csv('./btc_usd.csv', thousands=',')
```

```python
btc_to_usd['Date'] = pd.to_datetime(btc_to_usd['Date'], format='%m/%d/%Y')
```

```python
from matplotlib import pyplot as plt
```

```python
len(btc_to_usd)
```

```python
# the data is backwards, bruh [::-1]
plt.plot(btc_to_usd['Date'][:60], btc_to_usd['Open'][:60], color='#f73b31', label='Test data')
plt.plot(btc_to_usd['Date'][60:], btc_to_usd['Open'][60:], label='Train data')
plt.legend()
plt.xlabel('Date')
plt.ylabel('Price, USD')
plt.plot()
```
