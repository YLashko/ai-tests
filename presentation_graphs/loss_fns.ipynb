```python
import torch
```

```python
mse = torch.nn.MSELoss()
mae = torch.nn.L1Loss()
huber = torch.nn.HuberLoss()
```

```python
mse_list = []
mae_list = []
huber_list = []
for i in range(-400, 400, 2):
    x = torch.tensor([i / 200])
    mse_list.append(mse(torch.tensor([0.]), x))
    mae_list.append(mae(torch.tensor([0.]), x))
    huber_list.append(huber(torch.tensor([0.]), x))
```

```python
from matplotlib import pyplot as plt

plt.plot(list([i / 200 for i in range(-400, 400, 2)]), mse_list, color='red', label='MSE')
plt.plot(list([i / 200 for i in range(-400, 400, 2)]), mae_list, color='green', label='MAE')
plt.plot(list([i / 200 for i in range(-400, 400, 2)]), huber_list, color='blue', label='Huber')
plt.xlabel('Predicted value (target is 0)')
plt.ylabel('Calculated loss')
plt.legend()
plt.show()
```
