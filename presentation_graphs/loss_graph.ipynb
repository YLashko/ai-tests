```python
import numpy as np
from matplotlib import pyplot as plt

length = 100

loss_data = ((np.sin(np.array(list(range(length)) * np.linspace(1/15, 1/15, length)) + 0.3) * 100) + np.sin(np.array(list(range(length)) * np.linspace(1/4.2, 1/4.2, length))) * 40) + 130
```

```python
plt.plot(list(range(length)), loss_data)
plt.xlabel('Parameters combination')
plt.ylabel('Loss')
plt.show()
```
