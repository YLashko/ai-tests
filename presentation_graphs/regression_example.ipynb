```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
x = np.arange(1, 101).reshape(-1, 1)
y = np.linspace(1, 50, 100) + np.random.uniform(-5, 5, 100)
```

```python
model = LinearRegression()
y_flat = np.repeat(25, 100).reshape(-1, 1)
model.fit(x, y)
y_pred = model.predict(x)
print(model.coef_, model.rank_)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='#FF5733', label='Data points')
plt.plot(x, y_pred, color='#33FF57', label='Regression line')
plt.xlabel('X, Independent Variable')
plt.ylabel('Y, Dependent Variable')
plt.legend()
plt.savefig('./images/1_regression_example-2.png', bbox_inches='tight')
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='#FF5733', label='Data points')
plt.plot(x, y_flat, color='#33FF57', label='Regression line')
plt.xlabel('X, Independent Variable')
plt.ylabel('Y, Dependent Variable')
plt.legend()
plt.savefig('./images/1_regression_example-1.png', bbox_inches='tight')
```
